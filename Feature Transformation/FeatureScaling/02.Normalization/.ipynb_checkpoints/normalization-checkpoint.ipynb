#### Loading the required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Reading the data

```python
data = pd.read_csv('wine_data.csv', usecols = [0,1,2])
data.columns = ['Class label', 'Alcohol', 'Malic acid']
data.head()
```

```python
data.shape
```

```python
# visualizing the distribution of alcohol
sns.kdeplot(data['Alcohol'])
```

```python
# visualizing the distribution of malic acid
sns.kdeplot(data['Malic acid'])
```

```python
# plotting scatter plot
sns.scatterplot(data=data, x="Alcohol", y="Malic acid", hue="Class label", palette = 'deep',)
```

Doing train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('Class label',axis = 1),
                                                    data['Class label'],
                                                    test_size=0.3, 
                                                    random_state=0)
X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

# now, creating the object
scaler = MinMaxScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# converting the numpy array into the dataframe
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

```python
# comparing the before and after scaling datasets

fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))

ax1.scatter(X_train['Alcohol'], X_train['Malic acid'], c = y_train)
ax1.set_title("Before Scaling")

ax2.scatter(X_train_scaled['Alcohol'], X_train_scaled['Malic acid'], c = y_train)
ax2.set_title("After Scaling")
```

Through this scatterplot, we found that the data and its shape doesn't get changed even after scaling........... but after the scaling distribution of data gets changed and it is compressed inside the unit rectangle

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))

# before scaling
ax1.set_title("Before Scaling")
sns.kdeplot(X_train['Alcohol'], ax = ax1)
sns.kdeplot(X_train['Malic acid'], ax = ax1)

# after scaling
ax2.set_title("After Scaling")
sns.kdeplot(X_train_scaled['Alcohol'], ax = ax2)
sns.kdeplot(X_train_scaled['Malic acid'], ax = ax2)
plt.legend()
plt.show()
```

Here we observed that, before scaling Malic acid is little less than alcohol but after scaling we saw that before came in the same range i.e. between 0 to 1

#### Now, visualizing it individually

```python
# distribution of alcohol before and after scaling

fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))

# before scaling
ax1.set_title("Before Scaling")
sns.kdeplot(X_train['Alcohol'], ax = ax1)

# after scaling
ax2.set_title("After Scaling")
sns.kdeplot(X_train_scaled['Alcohol'], ax = ax2)

plt.show()
```

```python
# distribution of Malic acid before and after scaling

fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))

# before scaling
ax1.set_title("Before Scaling")
sns.kdeplot(X_train['Malic acid'], ax = ax1)

# after scaling
ax2.set_title("After Scaling")
sns.kdeplot(X_train_scaled['Malic acid'], ax = ax2)
plt.show()
```

From the both graph, we saw that distribution didn't get changed apart from their scale
